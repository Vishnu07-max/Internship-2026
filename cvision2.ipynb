```python
import ultralytics as ut
from ultralytics import YOLO
#downloads yolo11 offline
model = YOLO('yolo11n.pt')
predict_class = [0,14]
img = '.\lio2.jpg' 
pred = model(img , classes = predict_class, conf =.4)
pred_img = pred[0].plot()


import cv2
cv2.imwrite('image2.jpg', pred_img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2

# Load YOLO model
model = YOLO("yolov8n.pt")

# Image path
img_path = "Dog.jpg"   # Replace with your image name

# Detect only dogs (class ID = 16)
results = model(img_path, classes=[16], conf=0.25)

# Get annotated image
annotated_img = results[0].plot()

# Count dogs
dog_count = len(results[0].boxes)

print("Total Dogs Detected:", dog_count)

# Save output image
cv2.imwrite("detected_dogs.jpg", annotated_img)

# Show output
cv2.imshow("Dog Detection", annotated_img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```
