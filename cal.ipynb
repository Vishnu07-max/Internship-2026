```python
import math
from sklearn.feature_extraction.text import TfidfVectorizer
doc=["I love programming in Python.",
    "Python is a great programming language.",
    "I enjoy solving problems with Python."]
tfidf = TfidfVectorizer()
result = tfidf.fit_transform(doc)
print(tfidf.get_feature_names_out())
print(result)
```

```python
from nltk.tokenize import sent_tokenize
text = "Hello! How are you? I hope you're doing well. Let's learn some Python."
sent_tokenize(text)
```

```python
from nltk.stem import SnowballStemmer, LancasterStemmer
sentence = "I am enjoying learning Natural Language Processing with Python."
words = nltk.word_tokenize(sentence)
for word in words:
    print(SnowballStemmer("english").stem(word))
    print(LancasterStemmer().stem(word))
```
